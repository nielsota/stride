```python
import datetime
from rich import print

from stride.provider.strava.main import StravaService
from stride.stridedb.converters import StrideConverterService
```

```python
strava_service = StravaService()
```

```python
activities = strava_service.get_activities_by_date_range(
    start_date=datetime.datetime(2025, 8, 1),
    end_date=datetime.datetime(2025, 8, 10),
)
```

```python
activities
```

```python
streams = strava_service.get_streams(activity_id=activities[0].id)
print(streams)
```

```python
streams = [strava_service.get_streams(activity.id) for activity in activities]
print(streams)
```

```python
converted_activities = [StrideConverterService.process_strava_data(activity, streams) for activity, streams in zip(activities, streams)]

print(converted_activities)
```
